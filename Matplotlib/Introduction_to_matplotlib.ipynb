# Which one should  you use? (pyplot or matplotlib OO method)

* When plotting something quickly, okay to use pyplot method
* When plotting something more advanced, use the OO method

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
over_50.plot(kind='scatter',
            x = 'age',
            y = 'chol',
            c = 'target');
```

```python
# OO method mixed with pyplot method

fig, ax = plt.subplots(figsize=(10,5))

over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
```

```python
## OO method from scratch

fig, ax = plt.subplots(figsize = (10, 6))

scatter = ax.scatter(x = over_50['age'],
                    y = over_50['chol'],
                    c = over_50['target']);

ax.set(xlabel = 'age', ylabel='chol',title = 'heart_dieseas');

# Add legend

ax.legend(*scatter.legend_elements(), title = 'target');
```

```python
fig, ax = plt.subplots(figsize = (10,6))

scatter = ax.scatter(x = over_50['age'],
                    y = over_50['chol'],
                    c = over_50['target']);

ax.set(title = "Heart diesease chol level", xlabel = 'Age', ylabel = 'Chol');

ax.legend(*scatter.legend_elements(),title = 'Target');

# adding horizontal line

ax.axhline(y = over_50['chol'].mean(),
          linestyle='--');
```

```python
# creating subplots
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize = (10,10))

# adding data to subplots

scatter = ax0.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target']);

scatter1 = ax1.scatter(x = over_50['age'],
                      y = over_50['thalach'],
                      c = over_50['target']);

# adding legend to subplots

ax0.set(title='cholestrol level', xlabel = 'Age', ylabel= 'cholestrol');
ax0.legend(*scatter.legend_elements(), title='Target');

ax1.set(title = 'Highest heartrate', xlabel = 'Age', ylabel='Heartrate');
ax1.legend(*scatter.legend_elements(), title='Target');

# adding line to subplots

ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--');

ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--');

# Adding title to entire figure

fig.suptitle("Heart diesease", fontsize = 20, fontweight = 'bold');
```

```python
fig, ax = plt.subplots()
ax.bar(over_50['age'], over_50['chol']);

ax.set(title = 'cholestrol', xlabel='Age', ylabel='Chol level');

ax.axhline(y=over_50['chol'].mean(),
          linestyle = '--');
```

```python
# set the plote style
plt.style.use('seaborn-whitegrid')

# figure size
fig, ax = plt.subplots(figsize=(10, 6))

# plot the data
scatter = ax.scatter(x= over_50['age'],
                     y= over_50['chol'],
                     c= over_50['target'],
                     cmap = 'winter'); # it will change the color this is a color function 
# set the title
ax.set(title = 'Cholestrol',
       xlabel= 'Age', 
       ylabel='Chole');

# add the legend
ax.legend(*scatter.legend_elements(), 
          title='target')

# add the horizontal line
ax.axhline(y = over_50['chol'].mean(),
           linestyle = '--');
```

```python
# creating subplots
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize = (10,10))

# adding data to subplots
scatter = ax0.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'],
                     cmap = 'winter'); # adding custom colors

scatter1 = ax1.scatter(x = over_50['age'],
                      y = over_50['thalach'],
                      c = over_50['target'],
                      cmap = 'winter'); # adding custom colors


# seeting x and y axis limits
ax0.set_xlim([50, 80])
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# adding legend to subplots
ax0.set(title='cholestrol level', xlabel = 'Age', ylabel= 'cholestrol');
ax0.legend(*scatter.legend_elements(), title='Target');

ax1.set(title = 'Highest heartrate', xlabel = 'Age', ylabel='Heartrate');
ax1.legend(*scatter.legend_elements(), title='Target');


# adding line to subplots
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--');

ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--');


# Adding title to entire figure
fig.suptitle("Heart diesease", fontsize = 20, fontweight = 'bold');
```
